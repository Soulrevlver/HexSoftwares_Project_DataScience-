```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib import style
```

```python
iris = sns.load_dataset('iris')
```

```python
#Display the first 5 lines 
print(iris.head())
```

```python
#Check datset shape (rows, columns)
print("Dataset shape:", iris.shape)
```

```python
#Check coulmn names and data types 
print("\nDataset into:")
print(iris.info())
```

```python
#SUmmary statistics
print("\nSummary statistics:")
print(iris.describe())
```

```python
print("\nMissing values:")
print(iris.isnull().sum())
```

```python
#Check duplicates
print("Duplicates rows:", iris.duplicated().sum())
```

```python
#Drop duplicates if any 
iris = iris.drop_duplicates()
```

```python
#Confirm final structure
iris.info()
```

```python
# Explore class distributions
print("\n Count pf each species:")
print(iris['species'].value_counts())

sns.countplot(x='species', data=iris)
plt.title("Distribution of Iris Species")
plt.show()
          
```

```python
#Statistical insight per species
print("\n Mean values per spieces:")
print(iris.groupby('species').mean(numeric_only=True))

print("\n Variance per species:")
print(iris.groupby('species').var(numeric_only=True))
```

```python
#Pairplot to visualize relationship between features
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle('Iris Pairplot', y=1.02)
plt.show()
```

```python
#Correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(iris.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# boxplots for feature
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for col in features:
    plt.figure(figsize=(5,3))
    sns.boxplot(x='species', y=col, data=iris)
    plt.title(f'{col.capitalize()} Distributions by Species')
    plt.show()
```
